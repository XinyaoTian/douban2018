## 1、需要抓取的300部最新电影数据分析(http://movie.douban.com)
* 点击页面->**最近热门电影**里的**最新**->再点**更多**
* 页面下面有个JavaScript写的**加载更多**，点击一次，加载20条最新的记录，地址栏的page_start累加20
* 尝试将累加的地址直接复制到地址栏中看能不能用拼接方式获取所有页面，发现会自动跳回page_start=0，失败
* 那打开chrome的Inspect，分析一下网络数据，点击**加载更多**后，神奇的事儿发生了，返回了Json数据
* 直接复制返回Json数据的地址到浏览器地址栏：
** https://movie.douban.com/j/search_subjects?type=movie&tag=%E6%9C%80%E6%96%B0&page_limit=20&page_start=20 **
* 确实返回了我们需要的Json数据
* 于是确认采取页面拼接的方式抓取我们需要的所有电影的页面
#### 获取300条最新电影链接的方法：
1. 拼接所有页面，300/20 = 15个页面
2. 获取页面内容(json)
3. 将json导入到pandas，每获取一个json，就新导入到临时pandas中，再concat到第一个pandas的最后
4. 得到了一个pandas的数据表

```python
import pandas as pd
headers={'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.94 Safari/537.36'}
def get_urls():
    urls = []
    main_url = r'https://movie.douban.com/j/search_subjects?type=movie&tag=%E6%9C%80%E6%96%B0&page_limit=20&page_start='
    step = 20
    page_start = 0
    while page_start < 300:
        url = main_url + str(page_start)
        page_start += step
        urls.append(url)
    return urls
```

```python
import requests
import json

def get_df_data(url):
    html_json = requests.get(url,headers=headers).json()
    j = json.dumps(html_json)
    df = pd.read_json(j[13:-1],orient='records')
    return df
url = r'https://movie.douban.com/j/search_subjects?type=movie&tag=%E6%9C%80%E6%96%B0&page_limit=20&page_start=0'
df = get_df_data(url)
df
```

```python
import time,random
import requests
import json

def get_df_data(url):
    html_json = requests.get(url,headers=headers).json()
    j = json.dumps(html_json)
    df = pd.read_json(j[13:-1],orient='records')
    return df

def get_all_df_data(urls):
    df = get_df_data(urls[0])
    for url in urls[1:]:
        print(url)
        print(df[['title','url']].tail(1))
        time.sleep(random.random()*3)
        df = df.append(get_df_data(url))
    return df
urls = get_urls()
df = get_all_df_data(urls)
df
```

```python
df.reset_index(inplace=True)
drop_list = ['index']
df = df.drop(drop_list,axis=1)
df
```

```python
#获取所有链接：

def get_main_link(start_url):
    movie_id = start_url.split('/')[-2]
    main_str = 'https://movie.douban.com/subject/'+ movie_id + '/comments'
    return main_str

def get_html(url):
    html = session.get(url,headers=headers)
    return html

def get_next_page(next_click,t,u):
    comment_link = main_str + next_click
    
    html = get_html(comment_link)
    #找到next_click的拼接页面
    selector = etree.HTML(html.text)
    
    get_comments_content(selector,t,u)
    next_click = selector.xpath('//a[@class="next"]/@href')[0]
   
    
    return next_click

def get_text(xpath_list):
    if len(xpath_list) > 0:
        return xpath_list[0].strip()
    else:
        return ''

def get_user_info(user_link):
    html = session.get(url,headers=headers)
    selector = etree.HTML(html.text)
    location = get_text(selector.xpath('//div[@class="user-info"]/a/text()'))
    user_id_and_join_time = selector.xpath('//div[@class="user-info"]/div/text()')
    user_id = user_id_and_join_time[0].strip()
    join_time = user_id_and_join_time[1].strip()[:-2]
    
    return user_id,join_time,location


def get_comments_content(selector,t,u):
    comments = selector.xpath('//div[@class="comment"]')
    for comment in comments:
        try:
            short_comment = comment.xpath('.//p/text()')
            vote = comment.xpath('.//span[@class="votes"]/text()')
            user_link = comment.xpath('.//span[@class="comment-info"]/a/@href')
            user_name = comment.xpath('.//span[@class="comment-info"]/a/text()')
            seen = comment.xpath('.//span[@class="comment-info"]/span[1]/text()')
            rating = comment.xpath('.//span[@class="comment-info"]/span[2]/@class')
            
            comment_time = comment.xpath('.//span[@class="comment-info"]/span[3]/@title')
            
            #将user_link链接内的信息采集到
            #后来被封了，考虑暂时不采集用户信息，单独来采集
#             user_id,join_time,location = get_user_info(user_link)

            #print(user_link)
            
            #这儿的代码做了一个调整，就算是空，我也保留着：
            short_comment = short_comment[0].strip() if len(short_comment) > 0 else ''
            vote = vote[0].strip() if len(vote) > 0 else ''
            user_link = user_link[0].strip() if len(user_link) > 0 else ''
            user_name = user_name[0].strip() if len(user_name) > 0 else ''
            seen = seen[0].strip() if len(seen) > 0 else ''
            rating = rating[0][7:8] if len(rating) > 0 else ''
            comment_time = comment_time[0].strip() if len(comment_time) > 0 else ''
            
            
            
            short_comments.append(short_comment)
            votes.append(vote)
            user_links.append(user_link)
            user_names.append(user_name)
            seens.append(seen)
            ratings.append(rating)
            comment_times.append(comment_time)
            movies.append(t)
            movie_links.append(u)
            
#             users_id.append(user_id)
#             joins_time.append(join_time)
#             locations.append(location)
        except:
            continue
```
