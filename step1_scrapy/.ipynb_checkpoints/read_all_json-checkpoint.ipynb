```python
df_comments = frame[frame['title']=='全球风暴']['comment'].values
```

```python
comments = ''
for comment in df_comments:
    comments = comments + comment.strip()
    
#去掉标点符号
import re

pattern = re.compile(r'[\u4e00-\u9fa5]+')
filterdata = re.findall(pattern, comments)
cleaned_comments = ''.join(filterdata)
cleaned_comments[:500]

#词频统计
import jieba    #分词包

segment = jieba.lcut(cleaned_comments)
words_df=pd.DataFrame({'segment':segment})
stopwords=pd.read_csv("../step2_visuization/stopwords.txt",index_col=False,quoting=3,sep="\t",names=['stopword'], encoding='utf-8')#quoting=3全不引用
words_df=words_df[~words_df.segment.isin(stopwords.stopword)]
import numpy    #numpy计算包
words_stat = words_df.groupby('segment')['segment'].agg({'count':'size'}).reset_index().sort_values(by=["count"],ascending=False)

import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib
matplotlib.rcParams['figure.figsize'] = (10.0, 5.0)
from wordcloud import WordCloud#词云包

wordcloud=WordCloud(font_path="../step2_visuization/SimHei.ttf",background_color="white",max_font_size=80) #指定字体类型、字体大小和字体颜色
word_frequence = {x[0]:x[1] for x in words_stat.head(1000).values}
from PIL import Image
bg_pic = numpy.array(Image.open("./background.jpg"))
wordcloud = WordCloud(
        font_path="../step2_visuization/SimHei.ttf",
        background_color="white",
        max_font_size=80,
        width = 900,
        height = 900,
        mask = bg_pic,
        mode = "RGBA"
    )
wordcloud=wordcloud.fit_words(word_frequence)
plt.imshow(wordcloud)
```

```python
wordcloud.to_file('全球风暴.png')
```
